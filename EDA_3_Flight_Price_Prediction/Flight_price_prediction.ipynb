### We will look at these in a moment, lets deal with arrival and departure time first

```python
df['Arrival_Time'].str.split(" ").str[0]
```

```python
df['Arrival_Time'] = df['Arrival_Time'].apply(lambda x : x.split(' ')[0])
# removing the date 
```

```python
df['Arrival_hr'] = df['Arrival_Time'].apply(lambda x : int(x.split(':')[0]))
df['Arrival_min'] = df['Arrival_Time'].apply(lambda x : int(x.split(':')[1]))
# converting to min and hrs
```

```python
df.info()
```

```python
df['Dep_hr'] = df['Dep_Time'].apply(lambda x : int(x.split(':')[0]))
df['Dep_min'] = df['Dep_Time'].apply(lambda x : int(x.split(':')[1]))
# converting to min and hrs
```

```python
df.head()
```

```python
df.drop(['Arrival_Time', 'Dep_Time'], axis=1, inplace=True)
```
