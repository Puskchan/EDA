## Flight price prediction - EDA + FE

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df_train = pd.read_excel('data/Data_Train.xlsx')
df_test = pd.read_excel('data/Test_set.xlsx')
```

```python
df = pd.concat([df_train, df_test], axis=0)
```

```python
df.head()
```

```python
df.tail()
```

```python
df.info()
```

```python
df['Date_of_Journey'] = pd.to_datetime(df['Date_of_Journey'], dayfirst=True)
```

```python
df['Journey_Day'] = df['Date_of_Journey'].dt.day
df['Journey_Month'] = df['Date_of_Journey'].dt.month
df['Journey_Year'] = df['Date_of_Journey'].dt.year
df['Day_of_Week'] = df['Date_of_Journey'].dt.dayofweek
```

### Monday is 0 and Sunday is 6

```python
df.drop(['Date_of_Journey', 'Journey_Year'], axis=1, inplace=True)
```

### We will look at these in a moment, lets deal with arrival and departure time first

```python
df['Arrival_Time'].str.split(" ").str[0]
```

```python
df['Arrival_Time'] = df['Arrival_Time'].apply(lambda x : x.split(' ')[0])
# removing the date 
```

```python
df['Arrival_hr'] = df['Arrival_Time'].apply(lambda x : int(x.split(':')[0]))
df['Arrival_min'] = df['Arrival_Time'].apply(lambda x : int(x.split(':')[1]))
# converting to min and hrs
```

```python
df.info()
```

```python
df['Dep_hr'] = df['Dep_Time'].apply(lambda x : int(x.split(':')[0]))
df['Dep_min'] = df['Dep_Time'].apply(lambda x : int(x.split(':')[1]))
# converting to min and hrs
```

```python
df.head()
```

```python
df.drop(['Arrival_Time', 'Dep_Time'], axis=1, inplace=True)
```
